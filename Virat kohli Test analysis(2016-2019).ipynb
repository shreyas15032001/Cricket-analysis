```python
df = pd.read_csv('TESTS_ball_by_ball.csv')
```

```python
def player_out(player_dismissed):
    if type(player_dismissed) == str:
        return 1
    else:
        return 0
```

```python
df['Player_out'] = df.apply(lambda x: player_out(x['player_dismissed']), axis = 1)
```

```python
def is_bowler_wk(wicket_type, player_dismissed):
    if type(player_dismissed) == str:
        if wicket_type not in ('run out','hit wicket','retired hurt',0):
            return 1
        else:
            return 0
    
    else:
        return 0
```

```python
df['Bowler_wk'] = df.apply(lambda x: is_bowler_wk(x['wicket_type'], x['player_dismissed']), axis = 1)
```

```python
    df['is_dot'] = df.runs_off_bat.apply(lambda x: 1 if x == 0 else 0)
    df['is_one'] = df.runs_off_bat.apply(lambda x: 1 if x == 1 else 0)
    df['is_two'] = df.runs_off_bat.apply(lambda x: 1 if x == 2 else 0)
    df['is_three'] = df.runs_off_bat.apply(lambda x: 1 if x == 3 else 0)
    df['is_four'] = df.runs_off_bat.apply(lambda x: 1 if x == 4 else 0)
    df['is_six'] = df.runs_off_bat.apply(lambda x: 1 if x == 6 else 0)
```

```python
Virat_match_1 = Virat[Virat.match_id == 1022593]
```

```python
Virat_match_1 = Virat_match_1.sort_values(by = 'ball')
Virat_match_1['cumsum_balls'] = [i for i in range(1, len(Virat_match_1) + 1)]
```

```python
def Phases_of_play(cumsum_balls):
     if cumsum_balls < 25:
        return 'First 25 balls'
     elif cumsum_balls < 50:
        return 'Between 25-50 Balls'
     elif cumsum_balls < 75:
        return 'Between 50-75 Balls'
     elif cumsum_balls < 100:
        return 'Between 75-100 Balls'
     elif cumsum_balls < 125:
        return 'Between 100-125 Balls'
     elif cumsum_balls < 150:
        return 'Between 125-150 Balls'
     elif cumsum_balls < 175:
        return 'Between 150-175 Balls'
     elif cumsum_balls < 200:
        return 'Between 175-200 Balls'
     elif cumsum_balls < 225:
        return 'Between 200-225 Balls'
     elif cumsum_balls < 250:
        return 'Between 225-250 Balls'
     elif cumsum_balls < 275:
        return 'Between 250-275 Balls'
     elif cumsum_balls < 300:
        return 'Between 275-300 Balls'
     elif cumsum_balls < 325:
        return 'Between 300-325 Balls'
     elif cumsum_balls < 350:
        return 'Between 325-350 Balls'
     elif cumsum_balls < 375:
        return 'Between 350-375 Balls'
     elif cumsum_balls < 400:
        return 'Between 375-400 Balls'
     elif cumsum_balls < 425:
        return 'Between 400-425 Balls'
     elif cumsum_balls < 450:
        return 'Between 425-450 Balls'
     elif cumsum_balls < 475:
        return 'Between 450-475 Balls'
     elif cumsum_balls < 500:
        return 'Between 475-500 Balls'
```

```python
Virat_match_1['Phases'] = Virat_match_1.apply(lambda x: Phases_of_play(x['cumsum_balls']), axis = 1)
```

```python
Virat_match_1 = Virat_match_1[['match_id','innings','bowling_team','runs_off_bat','ball','is_dot','is_four','is_six','cumsum_balls','Phases','Player_out']]
```
